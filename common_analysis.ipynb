### Import Statement

```python
import pandas as pd
import numpy as np
import sys
import re
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
from plotly.offline import init_notebook_mode, iplot, plot
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt

# jupyter setup
init_notebook_mode(connected=True)
```

 ###  For the graph to be interative

```python
INTERATIVE_MODE = 0
if INTERATIVE_MODE:
    display_as = ''
else:
    display_as = 'png'
```

```python
df_2019 = pd.read_csv('dataset/developer_survey_2019/survey_results_public.csv')
df_2020 = pd.read_csv('dataset/developer_survey_2020/survey_results_public.csv')
languages = ['Python', 'JavaScript', 'Java', 'HTML/CSS', 'C++', 'C', 'C#', 'Go',
             'Bash/Shell/PowerShell', 'SQL', 'TypeScript', 'R', 'PHP', 'Ruby', 'Assembly', 'Rust',
             'VBA', 'WebAssembly', 'Other(s):' ,'Swift', 'Kotlin', 'Erlang', 'Elixir', 'Clojure','Objective-C',
            'Scala','F#','Dart', 'Perl', 'Julia','Haskell']
entries_in_2019 = len(df_2019)
entries_in_2020 = len(df_2020)
print(entries_in_2019, entries_in_2020)
```

### In the year 2019

```python
filt = df_2019.loc[:,'LanguageDesireNextYear'].notnull()
desired_languages_during_2019 = df_2019.loc[filt,'LanguageDesireNextYear']
```

```python
string_of_desired_languages_during_2019 = ''
for language in desired_languages_during_2019:
    splited_language_list = language.split(';')
    string_of_desired_languages_during_2019 += ' '.join(splited_language_list)
    string_of_desired_languages_during_2019 += ' '
```

```python
words = re.findall(r'\w+', string_of_desired_languages_during_2019.lower())
words_counter = Counter(words)
print(words_counter)
```

```python
wc = WordCloud(width = 900,height = 1500,max_words=15, background_color='#f5f5f5')
wc.generate_from_frequencies(words_counter)
wc.to_file("./output_images/want_to_learn_during_2019.png")
plt.subplots(figsize = (8,8))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show(display_as)
```

# Most Used Operating System  

```python
os_2019 = df_2019.loc[:,'OpSys'].value_counts()
os_2020 = df_2020.loc[:,'OpSys'].value_counts()
operating_system_df = pd.concat([os_2019, os_2020],axis=1).reset_index()
operating_system_df.columns = ['os', '2019', '2020']
```

```python
# operation_sysytem_df.rename(columns={'index':'OS','OpSys':'2019','OpSys':'2020'},inplace=True)
operating_system_df
```

```python
fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])
fig.add_trace(go.Pie(labels=operating_system_df['os'], values=operating_system_df['2019'], name='2019',title='2019'),1,1)
fig.add_trace(go.Pie(labels=operating_system_df['os'], values=operating_system_df['2020'], name='2020',title='2020'),1,2)
fig.update_layout(title_text="Operating System Preference")
fig.show(display_as)
```
