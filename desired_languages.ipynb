```python
import pandas as pd
import numpy as np
import sys
import re
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
from plotly.offline import init_notebook_mode, iplot, plot
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt

# jupyter setup
init_notebook_mode(connected=True)
```

```python
df_2019 = pd.read_csv('dataset/developer_survey_2019/survey_results_public.csv')
df_2020 = pd.read_csv('dataset/developer_survey_2020/survey_results_public.csv')
languages = ['Python', 'JavaScript', 'Java', 'HTML/CSS', 'C++', 'C', 'C#', 'Go',
             'Bash/Shell/PowerShell', 'SQL', 'TypeScript', 'R', 'PHP', 'Ruby', 'Assembly', 'Rust',
             'VBA', 'WebAssembly', 'Other(s):' ,'Swift', 'Kotlin', 'Erlang', 'Elixir', 'Clojure','Objective-C',
            'Scala','F#','Dart', 'Perl', 'Julia','Haskell']
entries_in_2019 = len(df_2019)
entries_in_2020 = len(df_2020)
print(entries_in_2019, entries_in_2020)
```

```python
filt = df_2019.loc[:,'LanguageDesireNextYear'].notnull()
desired_languages_during_2019 = df_2019.loc[filt,'LanguageDesireNextYear']
```

```python
string_of_desired_languages_during_2019 = ''
for language in desired_languages_during_2019:
    splited_language_list = language.split(';')
    string_of_desired_languages_during_2019 += ' '.join(splited_language_list)
    string_of_desired_languages_during_2019 += ' '
```

```python
words = re.findall(r'\w+', string_of_desired_languages_during_2019.lower())
words_counter = Counter(words)
print(words_counter)
```

```python
wc = WordCloud(width = 1200,height = 1800,max_words=15, background_color='#f5f5f5')
wc.generate_from_frequencies(words_counter)
wc.to_file("want_to_learn_during_2019.png")
plt.subplots(figsize = (8,8))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
